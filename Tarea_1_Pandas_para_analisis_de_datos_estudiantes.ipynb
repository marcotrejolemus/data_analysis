Tercero: Después nos debemos situar en la dirección donte esta el archivo o en su defecto darle el path completo de la localización del archivo

```python
# Tercero: Se importa la librería pandas, para el manejo de datos.
# Se le asigna el alias "pd", que es el estándar de
# codificación de la comunidad.

import pandas as pd

# Como los datos están en un archivo CSV, en lugar del nombre de
# archivo se utiliza la URL para acceso raw a los datos,
# que es lo mismo que leer un archivo que tenemos físicamente
# en nuestro equipo. Se declara una variable para almacenar la URL.

# Crear la variable para almacenar el archivo CSV en este caso datos_naciones_csv
datos_naciones_csv = '/content/drive/MyDrive/nations.csv'

# Se cargan los datos del archivo en la variable "datos_naciones_csv" a un
# DataFrame de pandas llamado "naciones", usando la función
# pd.read_csv(), usando como nombre de archivo la variable
# que almacena la URL para acceder a los datos.

naciones = pd.read_csv(datos_naciones_csv, encoding='latin-1')
```
